**1.Import Library**

```python
import matplotlib.pyplot as plt
import numpy as np
```

**2.Dataset**

```python
#dataset
stores =['Stores A','Stores B', 'Stores C']
male = [150,200, 180]
female= [120,230,170]

#dataquarter
q1_male=[150, 180, 160]
q1_female = [140,190,280]

q2_male=[170, 190, 175]
q2_female=[130,210, 160]

q3_male=[135,140,300]
q3_female=[120,100,200]

q4_male=[125, 290,280]
q4_female=[120,210,160]
```

3. Basic stacekd vertical bar chart

```python

x = np.arange(len(stores))
plt.bar(x,male, label='Male', color='blue')
plt.bar(x,female, bottom = male, label='female', color='red')

plt.title('population by gender and store')
plt.xlabel('stores')
plt.ylabel('population')
plt.xticks(x,stores)
plt.legend()
plt.show()
```

4. kustomisasi staced vertical bar chart

```python


x = np.arange(len(stores))
plt.bar(x,male, label='Male', color='blue')
plt.bar(x,female, bottom = male, label='female', color='red')

plt.title('sales transaction by store')
plt.xlabel('stores')
plt.ylabel('sales')
plt.xticks(x,stores)
plt.legend()

for i in range(len(x)):
    plt.text(x[i], male[i]/2, str(male[i]), ha='center', color='white')
    plt.text(x[i],male[i]+female[i]/2, str(female[i]), ha='center', color='black')
```



5.multiple staced vertical barchart


```python

width = 0.2
x = np.arange(len(stores))

plt.bar(x-width * 1.5, q1_male, label= 'q1_male', color='purple', width=width)
plt.bar(x-width * 1.5, q1_female, bottom=q1_male, label='q1_female', color='lightblue', width=width)

plt.bar(x+width/2, q2_male, label= 'q2_male', color='navy', width=width)
plt.bar(x+width/2, q2_female, bottom= q2_male, label='q2_female', color='grey', width=width)

plt.bar(x-width/2, q3_male, label= 'q3_male', color='lightpink', width=width)
plt.bar(x-width/2, q3_female, bottom=q3_male, label='q3_female', color='lightgreen', width=width)

plt.bar(x+width * 1.5, q4_male, label= 'q4_male', color='green', width=width)
plt.bar(x+width * 1.5, q4_female, bottom= q4_male, label='q4_female', color='aquamarine', width=width)

plt.title('Population by Gender and Store (Multiple Quarter)')
plt.xlabel('Stores')
plt.ylabel('Population')
plt.xticks(x, stores)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
