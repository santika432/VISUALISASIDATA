b. Kustomisasi Basic Bar Chart

```python
import matplotlib.pyplot as plt
#data
#data
Jurusan= [
    'Teknik informatika',
    'Sistem informasi',
    'Bisnis digital',
]
Jumlah_Mahasiswa = [120, 150, 80]

#Membuat Basic barchart
plt.bar(Jurusan, Jumlah_Mahasiswa, color=['blue', 'orange', 'red'])
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Jumlah Mahasiswa Per Jurusan')
plt.xticks(rotation=45)
for i, v in enumerate (Jumlah_Mahasiswa):
    plt.text(i, v+5, str(v), ha='center') #menambahkan label pada batang
plt.show()
```

c. Menambahkan Multiple Bar Chart

```python
import matplotlib.pyplot as plt
#data
#data
Jurusan= [
    'Teknik informatika',
    'Sistem informasi',
    'Bisnis digital',
]
Jumlah_Mahasiswa = [120, 150, 80]

#data tambahan untuk tahun yang berbeda

Jumlah_Mahasiswa_2024 = [110, 140, 95]

x=range(len(Jurusan)) #index untuk batang
width = 0.4
plt.bar(x, Jumlah_Mahasiswa, width, label='2023')
plt.bar([i+width for i in x], Jumlah_Mahasiswa_2024, width, label='2024', color='orange')

plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Jumlah Mahasiswa Per Jurusan(2023 vs2024)')
plt.xticks(rotation=45)
plt.xticks([i+width/2 for i in x], Jurusan) # mengatur label x=axis ditengah
plt.legend()
plt.show()
```

d. Basic Bar Chart untuk Melihat Pola dan Tren

```python
import matplotlib.pyplot as plt

Jurusan = [
    'Teknik informatika',
    'Sistem informasi',
    'Bisnis digital',
]

tahun = ['2019','2020', '2021', '2022', '2023']
data = {
    'Teknik informatika': [120, 130, 140, 150, 160],
    'Sistem informasi': [110, 120, 145, 165, 137],
    'Bisnis digital': [80, 90, 100, 110, 120],
}

width = 0.2
x = range(len(tahun))

plt.bar(x, data['Teknik informatika'], width, label='Teknik informatika', color='blue')
plt.bar([i + width for i in x], data['Sistem informasi'], width, label='Sistem informasi', color='orange')
plt.bar([i + width*2 for i in x], data['Bisnis digital'], width, label='Bisnis digital', color='red')

plt.xlabel('Tahun')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Tren Jumlah Mahasiswa Per Jurusan (2019-2023)')

plt.xticks([i + width for i in x], tahun)
plt.legend()
plt.show()
```
